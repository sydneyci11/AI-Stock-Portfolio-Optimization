```python
import os
import pandas as pd
import numpy as np
import re

# Change the working directory to where Excel files are stored containing stock information
directory_path = r"C:\Users\Sydney\Desktop\KCL\Undergraduate\Year3\Final Individual Project\Data_Collection\Status_of_TWSE"
os.chdir(directory_path)

start_year = 2010

# Loop for all the files
for year in range(10): 
    current_year = start_year + year
    for month in range(12):
        month += 1
        # Construct file name based on month and year
        if month <= 9:
            file_name = f"chk{current_year}0{month}" + ".xls"

        else:
            file_name = f"chk{current_year}{month}" + ".xls"
        
        # Check if the file exists before attempting to process it
        if not os.path.exists(file_name):
            print(f"File {file_name} does not exist. Skipping.")
            continue
        
        try:
            print(f"Now is processing the file {file_name}.")
            df = pd.read_excel(file_name, sheet_name=None)
            
        except Exception as e:
            print(f"Error reading {file_name}: {e}")
            continue
            
        # Handle a special case for March 2015
        if file_name == "chk201503.xls":
            sheet_7 = df.get("7.print")
        else:
            sheet_7 = df.get('7')
        
        # Initialize lists to store extracted data
        stock_code = []
        stock_name = []
        trading_vol  = []
        trading_val = []
        current_month_last_price = []
        previous_month_last_price = []
        monthly_average_price = []
        total_shares_outstanding = []
        shares_listed = []
        captical_amount = []
        
        # Iterate through each row of the sheet starting from the 5th row
        for i in range(4, (len(sheet_7))):
            # Extract the stock code and name
            
            num_str = str(sheet_7.iat[i, 0]).strip()
            match = re.match(r"(\d+)", num_str)
            
            if match:
                num = int(match.group(1))
                
                # Process only if the stock number is less than 35 or contains 'A'
                if num < 35 or 'A' in num_str:
                    
                    if 'A' in num_str:
                        stock_code.append(num_str)
                        stock_name.append(sheet_7.iat[i,1].replace('\n', ' '))
                        
                        if np.isnan(sheet_7.iat[i,6]):
                            trading_vol.append("-")
                        else:
                            trading_vol.append(sheet_7.iat[i,6])
                            
                        if np.isnan(sheet_7.iat[i,7]):
                            trading_val.append("-")
                        else:
                            trading_val.append(sheet_7.iat[i,7])
               
                        if np.isnan(sheet_7.iat[i,8]):
                            current_month_last_price.append("-")
                        else:
                            current_month_last_price.append(sheet_7.iat[i,8])
            
                        if np.isnan(sheet_7.iat[i,9]):
                            previous_month_last_price.append("-")
                        else:
                            previous_month_last_price.append(sheet_7.iat[i,9])
        
                        if np.isnan(sheet_7.iat[i,10]):
                            monthly_average_price.append("-")
                        else:
                            monthly_average_price.append(sheet_7.iat[i,10])
        
                        if np.isnan(sheet_7.iat[i,11]):
                            total_shares_outstanding.append("-")
                        else:
                            total_shares_outstanding.append(sheet_7.iat[i,11])
        
                        if np.isnan(sheet_7.iat[i,12]):
                            shares_listed.append("-")
                        else:
                            shares_listed.append(sheet_7.iat[i,12])
        
                        if np.isnan(sheet_7.iat[i,13]):  
                            captical_amount.append("-")
                        else:
                            captical_amount.append(sheet_7.iat[i,13])
                            
                else:
                    stock_code.append(num_str)
                    stock_name.append(sheet_7.iat[i,1].replace('\n', ' '))
                    
                    if np.isnan(sheet_7.iat[i,6]):
                        trading_vol.append("-")
                    else:
                        trading_vol.append(sheet_7.iat[i,6])
                
                    if np.isnan(sheet_7.iat[i,7]):
                        trading_val.append("-")
                    else:
                        trading_val.append(sheet_7.iat[i,7])
        
                    if np.isnan(sheet_7.iat[i,8]):
                        current_month_last_price.append("-")
                    else:
                        current_month_last_price.append(sheet_7.iat[i,8])
            
                    if np.isnan(sheet_7.iat[i,9]):
                        previous_month_last_price.append("-")
                    else:
                        previous_month_last_price.append(sheet_7.iat[i,9])
        
                    if np.isnan(sheet_7.iat[i,10]):
                        monthly_average_price.append("-")
                    else:
                        monthly_average_price.append(sheet_7.iat[i,10])
        
                    if np.isnan(sheet_7.iat[i,11]):
                        total_shares_outstanding.append("-")
                    else:
                        total_shares_outstanding.append(sheet_7.iat[i,11])
        
                    if np.isnan(sheet_7.iat[i,12]):
                        shares_listed.append("-")
                    else:
                        shares_listed.append(sheet_7.iat[i,12])
        
                    if np.isnan(sheet_7.iat[i,13]):  
                        captical_amount.append("-")
                    else:
                        captical_amount.append(sheet_7.iat[i,13])

            else:
                print(f"Cannot process num: {num_str}")
            
        stock_status = pd.DataFrame({
            "股票代碼" : stock_code,
            "股票名稱" : stock_name, 
            "本月成交量(千股)" : trading_vol,
            "本月成交值(千元)" : trading_val,
            "本月底市價(元)" : current_month_last_price,
            "上月底市價(元)" : previous_month_last_price,
            "本月平均價(元)" : monthly_average_price,
            "發行股份(千股)" : total_shares_outstanding, 
            "上市股份(千股)" : shares_listed,
            "資本額(百萬元)" : captical_amount,
            })
        
        try:
            csv_file_name = f"{file_name[3:-4]}.csv"
            stock_status.to_csv(csv_file_name, index=False, encoding='utf-8-sig', sep=',')
        
            print(f"Save {csv_file_name} successfully.")
            
        except Exception as e:
            print(f"Error processing {file_name}: {e}")
            continue
```
